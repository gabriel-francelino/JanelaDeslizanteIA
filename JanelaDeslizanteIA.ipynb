# Matriz de Conhecimento
Matriz de conhecimento utilizando a técnica de janela deslizante de bebês nomeados de Gabriel, Emma e Tesla de 1880 até 2015.

### Importações das bibliotecas


```python
# importa as bibliotecas que serão usadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

### Janela deslizante para o nome Gabriel

```python
# ler a base de dados usada.
filename='https://raw.githubusercontent.com/gabriel-francelino/JanelaDeslizanteIA/main/names_over_time/gabrielOverTime.csv'
data = pd.read_csv(filename)

# Definir o tamanho da janela
window_size = 10

# Cria uma matriz sem inicializar as entradas
matrix = np.empty((0, window_size + 1))

# Cria a matriz da janela deslizante
for i in range(window_size, len(data)):
    window = data['Number'][i-window_size:i]
    target = data['Number'][i]
    matrix = np.vstack((matrix, np.append(window, target)))

# Cria um datatframe a partir da 'matrix_df' e rotula as colunas com os nomes de "Quantidade (i+1)" até "Saída"
matrix_df = pd.DataFrame(matrix, columns=[f'   Quantidade  {i+1}' for i in range(window_size)] + ['Saída'])

# Mostra a matriz
matrix_df
```

```python
# Separando os dados em duas listas para plotagem
x = data["YearOfBirth"]
y = data["Number"]

# Configurando o tamanho do gráfico
fig, ax = plt.subplots(figsize=(10, 5))

# Adicionando título e rótulos aos eixos
ax.set_title("Número de pessoas com o nome Gabriel por ano")
ax.set_xlabel("Ano de nascimento")
ax.set_ylabel("Número de pessoas")

# Plotando o gráfico de linha
ax.plot(x, y, color='blue')

# Exibindo o gráfico
plt.show()
```
